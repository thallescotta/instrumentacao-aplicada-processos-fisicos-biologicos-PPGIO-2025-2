🔹 Célula 2 — Parse, soma das áreas por banda/fluência e exemplo de gráfico

Esta célula usa a variável CLIPBOARD da célula anterior, monta uma tabela “longa” e plota Área × Fluência.

```python
#@title Parse do Clipboard → Tabela (banda × fluência × área) e gráfico Área×Fluência
import re
import math
import pandas as pd
import matplotlib.pyplot as plt

raw = CLIPBOARD

# --- helpers ---
def to_float_br(x: str) -> float:
    x = x.strip().replace("\u00A0", " ")  # nbsp
    # números vêm com vírgula decimal no clipboard
    x = x.replace('.', '').replace(',', '.') if re.search(r'\d,\d', x) else x.replace(',', '.')
    try:
        return float(x)
    except:
        return math.nan

# mapeamento de janelas → rótulo do pico (cm⁻¹)
WINDOW_LABELS = {
    (580.0, 592.0): "589",
    (1155.0, 1178.0): "1166",
    (1280.0, 1305.0): "1295",
    (2224.0, 2243.0): "2239",
    (2565.0, 2600.0): "2581",
    (3376.0, 3382.0): "3380",
    (3704.5, 3706.5): "3706",
}

def label_from_window(lo, hi):
    for (a,b), lab in WINDOW_LABELS.items():
        # tolerância de 0.2 cm-1
        if abs(lo-a) < 0.2 and abs(hi-b) < 0.2 or abs(lo-b) < 0.2 and abs(hi-a) < 0.2:
            return lab
    # fallback: usa centro da janela para achar o mais próximo
    center = 0.5*(lo+hi)
    candidates = {abs(center - float(k)): k for k in ["589","1166","1295","2239","2581","3380","3706"]}
    return candidates[min(candidates.keys())]

# --- parse em blocos ---
blocks = re.split(r'\bPEAKS RESOLVED:\s*', raw)[1:]  # ignora cabeçalho
rows = []

for blk in blocks:
    # Spectrum (para fluência)
    m_spec = re.search(r'Spectrum:\s*(.+)', blk)
    spectrum = m_spec.group(1).strip() if m_spec else "NA"
    # fluência = última "palavra" do Spectrum (ex.: 2.5e10, 1e12)
    flu = spectrum.split()[-1]

    # Region
    m_reg = re.search(r'Region:\s*([0-9,\.]+)\s+([0-9,\.]+)', blk)
    if not m_reg:
        continue
    hi, lo = m_reg.group(1), m_reg.group(2)  # no OMNIC o primeiro é o maior
    lo_f, hi_f = to_float_br(lo), to_float_br(hi)
    band = label_from_window(lo_f, hi_f)

    # linhas de picos → última coluna é Area
    area_vals = []
    for line in blk.splitlines():
        if re.search(r'\bVoigt\b', line) and re.search(r'\bArea\b', blocks[0]) or '\t' in line:
            # pega o último token numérico da linha
            parts = re.split(r'\s+', line.strip())
            # a linha vem tabulada; tentamos extrair o último campo
            try:
                last = parts[-1]
                val = to_float_br(last)
                if math.isfinite(val):
                    area_vals.append(val)
            except:
                pass

    area_sum = sum(area_vals) if area_vals else math.nan
    rows.append({
        "fluencia": flu,
        "spectrum": spectrum,
        "janela_lo": lo_f,
        "janela_hi": hi_f,
        "pico_cm-1": band,
        "area_abs_cm-1": area_sum
    })

df = pd.DataFrame(rows).sort_values(["pico_cm-1","fluencia"])
print("Resumo (long):")
display(df)

print("\nTabela (largura) — linhas: fluências, colunas: picos (área):")
wide = df.pivot_table(index="fluencia", columns="pico_cm-1", values="area_abs_cm-1", aggfunc="sum")
display(wide)

# --- gráfico Área × Fluência (log10 do eixo-x se houver formato tipo 2.5e10/1e12) ---
def flu_str_to_float(s):
    try:
        return float(s)
    except:
        # tenta converter 1e12, 2.5e10 etc.
        m = re.fullmatch(r'([0-9]*\.?[0-9]+)[eE]([+-]?\d+)', s)
        if m:
            return float(m.group(1)) * (10 ** int(m.group(2)))
        return math.nan

df_plot = df.copy()
df_plot["fluencia_num"] = df_plot["fluencia"].map(flu_str_to_float)

plt.figure(figsize=(7,5))
for band, g in df_plot.groupby("pico_cm-1"):
    plt.plot(g["fluencia_num"], g["area_abs_cm-1"], marker='o', label=f'{band} cm⁻¹')
plt.xscale('log')
plt.xlabel('Fluência (proj·cm⁻²) [escala log]')
plt.ylabel('Área integrada (abs·cm⁻¹)')
plt.title('Área × Fluência — bandas “Pai”')
plt.legend()
plt.grid(True, which='both', ls=':')
plt.show()
```
